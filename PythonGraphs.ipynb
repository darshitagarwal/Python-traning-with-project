```python
import numpy as np
import random
import matplotlib.pyplot as plt
class subplots:
    def graphplot():
        x_axis=np.random.randint(10,50,8)
        y_axis=np.random.randint(10,50,8)
        plt.subplot(2,2,1)
        # subplot(<no.of columns>,<Which column>,<which row in that column>)
        plt.grid(color="green")
        plt.plot(x_axis,marker="o",mfc="red",color="blue")
        plt.plot(y_axis,marker="o",mfc="purple",color="blue")

        a_axis=np.random.randint(10,40,8)
        b_axis=np.random.randint(10,30,8)
        plt.subplot(2,2,2)
        plt.grid()
        plt.barh(a_axis,b_axis)

        c_axis=np.random.randint(50,100,12)
        d_axis=np.random.randint(10,60,12)
        plt.subplot(2,2,3)
        plt.grid()
        plt.scatter(a_axis,b_axis)

        c_axis=np.random.randint(50,100,12)
        d_axis=np.random.randint(10,60,12)
        plt.subplot(2,2,4)        
        plt.pie(c_axis,labels=d_axis,shadow="True")
        plt.legend()
        plt.show()
    graphplot
subplots.graphplot()
subplots




```

```python
import random
class DataPlot:
    def PlotCoordinates():
        global x_axis,y_axis
        x_axis=np.random.randint(10,90,10)
        y_axis=np.random.randint(20,100,10)
        print("\t\t-----Coordinates-----")
        print("X-axis => ", x_axis)
        print("Y-axis => ", y_axis)
    def PlotGraph():
        plt.plot(x_axis,color="blue",marker="o",mfc="yellow",linestyle="dashed")
        plt.plot(y_axis,color="red",marker="o",mfc="black")
        plt.grid(color="green")
        plt.show()
DataPlot.PlotCoordinates()
DataPlot.PlotGraph()
```
